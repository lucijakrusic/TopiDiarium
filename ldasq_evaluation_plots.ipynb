Plotting topic terms

```python
def plot_topic_terms(ldaseq_model, time_slice_index, num_topics=5, num_terms=10):
    fig, axes = plt.subplots(num_topics, 1, figsize=(15, num_topics * 5), sharex=True)
    fig.suptitle(f'Top Terms in Each Topic for Time Slice {time_slice_index}', fontsize=16)

    for topic_id in range(num_topics):
        topic_terms = ldaseq_model.print_topic(topic_id, time=time_slice_index, top_terms=num_terms)
        terms = [term.split('*')[1].replace('"', '').strip() for term in topic_terms.split(' + ')]
        weights = [float(term.split('*')[0]) for term in topic_terms.split(' + ')]

        ax = axes[topic_id]
        ax.barh(terms, weights, color='blue')
        ax.set_title(f'Topic {topic_id}')
        ax.set_xlabel('Weight')
        ax.set_ylabel('Term')
        ax.invert_yaxis()  # Highest weights at the top

    plt.tight_layout(rect=[0, 0, 1, 0.96])
    plt.show()
```

```python
import os
import matplotlib.pyplot as plt
from gensim.corpora import Dictionary, MmCorpus
from gensim.models import LdaSeqModel
import logging

```
